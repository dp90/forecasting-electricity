```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('./data/energy_dataset.csv')
```

1. Missing Values and Duplicates:

Look for any missing values or duplicates in your data, which could skew your results.

```python
# Check for duplicates
print(f"Number of duplicates: {df.duplicated().sum()}")

# Check for missing values
print(df.isnull().sum())
```

2. Feature Distributions:

Plot histograms or box-plots of the feature distributions. This can reveal a lot about the underlying patterns in your data, the presence of outliers, etc.

```python
# Display the first few rows of the DataFrame
print(df.head())

# Print a summary of the DataFrame
print(df.info())

# Print descriptive statistics
print(df.describe())
```

```python
# Plot histogram for all variables
df.hist(bins=50, figsize=(20,15))
plt.show()
```

3. Correlation Analysis:

Understand how electricity demand and price are related to each other, and to other factors:

Correlation Coefficients: Use these to understand the linear relationship between variables.
Cross-Correlation: For time series data, it's often useful to measure the correlation between two series offset in time.

```python
# Plot pairwise relationships in the dataset
sns.pairplot(df)
plt.show()

# Compute and print correlation matrix
correlation_matrix = df.corr()
print(correlation_matrix)
```

4. Time-series Analysis:

Since electricity demand and price usually depend on the time of day, week, month, and year, make sure to consider these aspects:

Trends: Is there a long-term increase or decrease seen in the data?
Seasonality: Are there repeating short-term cycles in the data?
Outliers: Are there extreme values that deviate from the rest of the data points?
Periodicity: The occurrence of repeated patterns at regular intervals.
